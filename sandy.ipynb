# reading data:

```
#reading data
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
a=pd.read_csv("/results.csv")
print(a)
print(a.shape)
print(a.dtypes)

```

```
a.std(ddof=1)
```

```
a.describe()
```

```
a.var(ddof=1)
```

```
a.mean()
```

![alt text](https://)

```
a.isna().sum()
```

```
df=a.fillna(0)
```

```
print(sns.pairplot(data = df))
```

```

sns.relplot(x='raceId',y='number',hue='resultId',data=df)
```

```
count_pop = df.groupby('resultId')['constructorId'].mean()
count_pop.plot(kind = 'hist',)
plt.show()
```

```
position_carter=df.groupby('position')['grid'].mean()
position_carter.plot(kind='pie',)
plt.show()
```

```
point_rate=df.groupby('laps')['points'].mean()
point_rate.plot()
plt.xlabel('laps')
plt.ylabel('points')
plt.title('point rate')
plt.show()
```

```
raceId_carter = df.groupby('raceId')['positionOrder'].mean()
plt.style.use('ggplot')
raceId_carter.plot(kind='bar',)
plt.show()
```

```
sns.lineplot(x='grid',y='number',hue='grid',data=df)
```

```
sns.violinplot(x='points',y='number',cut=2,hue='points',data=df)
```

```
sns.catplot(x='statusId',y='time',hue='driverId',data=df)
```
